```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
matrizes = {
    "Regressão Logística": [[126420, 55411],
                            [9652, 11874]],

    "Random Forest": [[122032, 59799],
                      [10152, 11374]],

    "MLP": [[128547, 53284],
            [10033, 11493]],

    "SVM": [[125340, 56491],
            [9605, 11921]],

    "Árvore de Decisão": [[0, 181831],
                          [0, 21526]]
}

fig, axes = plt.subplots(3, 2, figsize=(12, 16))
axes = axes.flatten()

classes = ["Adimplente", "Inadimplente"]

for ax, (modelo, matriz) in zip(axes, matrizes.items()):
    sns.heatmap(
        matriz,
        annot=True,
        fmt="d",
        cmap="Blues",
        cbar=False,
        xticklabels=classes,
        yticklabels=classes,
        ax=ax,
        square=True
    )

    ax.set_title(modelo, fontsize=16)
    ax.set_xlabel("Previsto", fontsize=14)
    ax.set_ylabel("Real", fontsize=14)
    ax.tick_params(axis="x", labelsize=14)
    ax.tick_params(axis="y", labelsize=14)

# Remove o sexto gráfico vazio
axes[-1].axis("off")

plt.suptitle("", fontsize=18)
plt.tight_layout()
plt.savefig("./figuras/matrizes_confusao_modelos.png", dpi=300, bbox_inches="tight")
plt.show()
```
